# For the Topic 'Creativity'

```python

creativity = ["art", "positive", "attitudes", "technologies", "participants", "variables", "field", "results", "culture", "visual", "studies", "positively", "technology", "study", "autonomy", "general", "measured", "relatedness", "psychological", "people"]
creativity_len = len(creativity)
creativity_dict = {}
for i in range(creativity_len):
	count = 0
	with open("papers_text\papers.txt", encoding="utf8") as f: 
		for line in f: 
			words = line.split() 
			for j in words: 
				if(j==creativity[i]): 
					count=count+1
	creativity_dict[creativity[i]] = count
	
print(creativity_dict)

```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(creativity_dict)

# Display the generated word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# Open the file in read mode 
text = open("papers_text\papers.txt", encoding="utf8") 


# Download the stopwords dataset
nltk.download('stopwords')
nltk.download('punkt')

def remove_stopwords(text):
    # Tokenize the input text
    words = word_tokenize(text)

    # Get the English stopwords set from nltk
    stop_words = set(stopwords.words('english'))

    # Remove stopwords from the tokenized words
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Join the filtered words back into a string
    filtered_text = ' '.join(filtered_words)

    return filtered_text



output_text = remove_stopwords(text)


# Create an empty dictionary 
d = dict() 

# Loop through each line of the file 
for line in text: 
	# Remove the leading spaces and newline character 
	line = line.strip() 

	# Convert the characters in line to 
	# lowercase to avoid case mismatch 
	line = line.lower() 

	# Split the line into words 
	words = line.split(" ") 
						

	# Iterate over each word in line 
	for word in words: 
		# Check if the word is already in dictionary 
		if word in d: 
			# Increment count of word by 1 
			d[word] = d[word] + 1
		else: 
			# Add the word to dictionary with count 1 
			d[word] = 1

# Print the contents of dictionary 
for key in list(d.keys()): 
	print(key, ":", d[key]) 
```

```python
# Sample dictionary (replace this with your own dictionary)

# Get the top 20 words and their counts
top_words = sorted(d.items(), key=lambda x: x[1], reverse=True)[:20]

# Print the top 20 words and their counts
for word, count in top_words:
    print(f"{word}: {count}")
```
